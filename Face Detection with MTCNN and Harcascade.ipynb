# Detecting faces from images

```python
import cv2

cap = cv2.VideoCapture(0)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.imshow('Face Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Assignment2
# initialize the MTCNN face detection model
detector = MTCNN()

# initialize the video capture object using a video file
cap = cv2.VideoCapture('video.mp4')

# loop through the frames of the video
while True:
    # read the current frame from the video capture object
    ret, frame = cap.read()

    # if the frame is not read properly, break the loop
    if not ret:
        break
        
    # detect faces in the current frame using the MTCNN model
    faces = detector.detect_faces(frame)
    
    # loop through each detected face
    for face in faces:
        # extract the bounding box coordinates of the face
        x, y, width, height = face['box']
        
        # draw a rectangle around the face
        cv2.rectangle(frame, (x, y), (x+width, y+height), (0, 255, 0), 2)
        
    # show the current frame with the detected faces
    cv2.imshow('Face Detection', frame)
    
    # wait for a key press to exit
    if cv2.waitKey(1) == ord('q'):
        break

# release the video capture object and destroy all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Load the mask cascade
mask_cascade = cv2.CascadeClassifier("mask_cascade.xml")

# Load the image
#img = cv2.imread("unmask_person.jpeg")
img = cv2.imread("mask_image.jpeg")
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect the faces with masks
masks = mask_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Draw the rectangle around each face with mask
for (x, y, w, h) in masks:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display the output
cv2.imshow("Mask Detection", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
